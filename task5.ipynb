```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from scipy import sparse
```

```python
adjacency_matrix = np.zeros((33, 33))
adjacency_matrix[0][1] = 1
adjacency_matrix[1][2] = 1
adjacency_matrix[2][3] = 1
adjacency_matrix[3][4] = 1
adjacency_matrix[4][5] = adjacency_matrix[4][10] =  adjacency_matrix[4][32]= 1
adjacency_matrix[5][6] = 1
adjacency_matrix[6][7] = adjacency_matrix[6][8] = 1
adjacency_matrix[8][9] = 1
adjacency_matrix[10][11] = 1
adjacency_matrix[11][12] = 1
adjacency_matrix[12][13] = 1
adjacency_matrix[14][32] = adjacency_matrix[14][15] = adjacency_matrix[14][16] = 1
adjacency_matrix[15][32] = 1
adjacency_matrix[16][17] = adjacency_matrix[16][20] = adjacency_matrix[16][21] = adjacency_matrix[16][22] = 1
adjacency_matrix[17][18] = adjacency_matrix[17][19] = 1
adjacency_matrix[21][22] = 1
adjacency_matrix[23][24] = 1
adjacency_matrix[24][25] = adjacency_matrix[25][26] = 1
adjacency_matrix[26][27] = adjacency_matrix[26][29] = adjacency_matrix[26][31] = 1
adjacency_matrix[27][28] = 1
adjacency_matrix[29][30] = 1
adjacency_matrix += adjacency_matrix.transpose() 
plt.imshow(adjacency_matrix)









```

```python
# Создаем граф
G = nx.from_numpy_array(adjacency_matrix)

# Рисуем граф с настройками
pos = nx.spring_layout(G, k=0.35) # задаем расположение вершин
nx.draw(G, pos, with_labels=True, node_size=500, font_size=10) # задаем параметры вершин и меток
edge_labels = nx.get_edge_attributes(G, "weight") # получаем метки ребер из матрицы смежности
# nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels) # добавляем метки ребер
plt.show()
```

```python
incidence_matrix = sparse.csr_matrix.toarray(nx.incidence_matrix(G))
print((incidence_matrix))
plt.imshow(incidence_matrix)
```
