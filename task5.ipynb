```python
adjacency_matrix = np.zeros((28, 28))
adjacency_matrix[1][2] = 1
adjacency_matrix[2][3] = 1
adjacency_matrix[3][5] = 1
adjacency_matrix[4][7] = 1
adjacency_matrix[5][22] = 1
adjacency_matrix[6][7] = 1
adjacency_matrix[7][9] = adjacency_matrix[7][10] = adjacency_matrix[7][11] = adjacency_matrix[7][12] = 1
adjacency_matrix[8][9] = 1
adjacency_matrix[9][10] =adjacency_matrix[9][11] = adjacency_matrix[9][12] = 1
adjacency_matrix[10][11] = adjacency_matrix[10][12] = 1
adjacency_matrix[11][12] = 1
adjacency_matrix[13][14] = adjacency_matrix[13][15] = adjacency_matrix[13][16] = adjacency_matrix[13][18]= adjacency_matrix[13][20] = adjacency_matrix[13][21]= 1
adjacency_matrix[14][15] = adjacency_matrix[14][16] = adjacency_matrix[14][18] = adjacency_matrix[14][20]= adjacency_matrix[13][21] = 1
adjacency_matrix[15][16] = adjacency_matrix[15][18] = adjacency_matrix[15][20]= adjacency_matrix[15][21] = 1
adjacency_matrix[16][17]= adjacency_matrix[16][18]= adjacency_matrix[16][20] = adjacency_matrix[16][21] =  adjacency_matrix[16][23]= 1
adjacency_matrix[18][20] = adjacency_matrix[18][21] = 1
adjacency_matrix[19][20]  = 1
adjacency_matrix[20][21] = 1
adjacency_matrix[22][24] = 1
adjacency_matrix[24][25] = adjacency_matrix[24][26] = 1
adjacency_matrix[26][27] = 1

adjacency_matrix += adjacency_matrix.transpose()
plt.imshow(adjacency_matrix)
```

```python
# Создаем граф
G = nx.from_numpy_array(adjacency_matrix)

# Рисуем граф с настройками
pos = nx.spring_layout(G) # задаем расположение вершин
nx.draw(G, pos, with_labels=True, node_size=500, font_size=10) # задаем параметры вершин и меток
edge_labels = nx.get_edge_attributes(G, "weight") # получаем метки ребер из матрицы смежности
# nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels) # добавляем метки ребер
plt.show()
```
